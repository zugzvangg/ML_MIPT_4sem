```python
 import pandas as pd
```

```python
data = pd.read_csv('Market_Basket_Optimisation.csv', header=None)
```

```python
data.head()
```

```python
from apyori import apriori
```

```python
data.fillna(method='ffill', axis = 1, inplace=True)
```

```python
data.head()
```

```python
dt = data.values.tolist() #Преобразование аналогичное тому, что в примере, но быстрее
```

```python
result = list(apriori(dt, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2))
```

```python
ls = []
for i in list(result):
    items1 = list(list(i)[0])
    items = ''
    for j in items1:
        items+=j
        items+=', '
    items = items[:-2]
    tmp = list(list(i[2])[0])
    items_base1 = list(list(tmp[0]))
    items_base = ''
    for j in items_base1:
        items_base+=j
        items_base+=', '
    items_base = items_base[:-2]
    items_add1 = list(tmp[1])
    items_add = ''
    for j in items_add1:
        items_add+=j
        items_add+=', '
    items_add = items_add[:-2]
    ls.append([items, items_base, items_add, round(tmp[2], 3), round(tmp[3],3), round(i[1], 5)])
```

```python
final = pd.DataFrame(ls, columns=['items', 'items_base', 'items_add', 'confidence', 'lift', 'support'])
```
